```python
import pandas as pd # standard convention

# Comparison Dictionary vs. Pandas dataframe

people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
print(type(people))
print(people)
print("\n")

people_df = pd.DataFrame(people)
print(type(people_df))

print(people_df)
people_df # Jupyter displays dataframes nicer without utilizing print
```

```python
# Another example of making Pandas Dataframe from python dictionary

countries_df = pd.DataFrame({
'country': ['United States', 'The Netherlands', 'Spain', 'Mexico', 'Australia'],
'capital': ['Washington D.C.', 'Amsterdam', 'Madrid', 'Mexico City', 'Canberra'],
'continent': ['North America', 'Europe', 'Europe', 'North America', 'Australia'],
'language': ['English', 'Dutch', 'Spanish', 'Spanish', 'English']})

countries_df
```
